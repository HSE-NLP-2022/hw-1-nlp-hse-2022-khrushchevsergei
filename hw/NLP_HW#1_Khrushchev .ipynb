#### Предобработаем наш текст
Для этого убираем пунктуацию, стоп слова, форматируем регистр, добивим формализацию через part-of-speech tagging. Это позволяет лучше обучаться модели 

https://www.analyticsvidhya.com/blog/2021/11/an-introduction-to-stemming-in-natural-language-processing/

```python
def tagging(term):
    label = nltk.pos_tag([term])[0][1][0].upper()
    dict_pos_tag = {"ADJ": wordnet.ADJ,
                    "NOUN": wordnet.NOUN,
                    "VERB": wordnet.VERB,
                    "ADV": wordnet.ADV}

    return dict_pos_tag.get(label, wordnet.NOUN)
```

```python
def purify(doc):
    result = []
    content = " ".join([line.lower().strip() for line in open(path + doc)])
    phrases = sent_tokenize(content)

    for phrase in phrases:
        phrase = re.sub("[^a-zA-Z]", " ", phrase)
        terms = [lemmatizer.lemmatize(term, 'v') for term in [lazer.lemmatize(term, tagging(term)) for term in phrase.strip().split()] if term not in stoppers and len(term) > 1]
        result.append(terms)
    
    return result
```

Для выполнения данного пункта напишем генератор и обучим модель Word2Vec

```python
def gen():
    for phrase in volume['phrases']:
        yield phrase.split()
```

```python
w2v = word2vec.Word2Vec(workers=10, size=200, min_count=2, window=6, sample=1e-3)
w2v.build_vocab(gen())
w2v.train(gen(), total_examples=w2v.corpus_count, epochs=15)
```

```python
# Delete stop words 
def exclude_stoppers(phrase):
    altered = regexp.tokenize(phrase)
    dumped = [term for term in altered if not term in stoppers and len(term) > 1]
    erased_list = ' '.join(dumped)
    return erased_list
```

```python
corpus_train['refined'] = corpus_train[2].apply(exclude_stoppers)
corpus_val['refined'] = corpus_val[2].apply(exclude_stoppers)
corpus_test['refined'] = corpus_test[2].apply(exclude_stoppers)
x_tr = corpus_train.refined.tolist()
y_tr = corpus_train[1].tolist()
```
